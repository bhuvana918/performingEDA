```python
df=df.drop_duplicates()
df.head(5)
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(df['volatile acidity'],color='green')
```

```python
df.chlorides.value_counts().nlargest(10).plot(kind='bar', figsize=(10,5),color='red')
plt.xlabel('Chlorides')
plt.ylabel('Wine Quality')
plt.title('Wine Quality on chlorides')
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['citric acid'],color='pink',edgecolor='black')
plt.title('Count of Citric Acid')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['fixed acidity'],y=df['pH'],color='m')
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x=df['free sulfur dioxide'],y=df['quality'],color='cyan')
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,annot=True)
```
